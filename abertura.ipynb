```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df['Postal Code'].fillna(df['Postal Code'].mean(), inplace=True)
```

```python
df.head(2)
```

```python
receita = df['Sales'].sum()
receita
```

```python
receita_estado = df.groupby('State')['Sales'].sum()
receita_estado
```

```python
plt.figure(figsize = (50, 18))
sns.barplot(x = receita_estado.index, y = receita_estado.values)
plt.title('Receita por Estado')
plt.xlabel('Estado')
plt.ylabel('Receita')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

df['Time'] = (df['Ship Date'] - df['Order Date']).dt.days
df[['Order Date', 'Ship Date', 'Time']].head()
```

```python
plt.figure(figsize = (10, 6))
sns.histplot(df['Time'], bins = 20, kde = True)
plt.title('DistribuiÃ§Ã£o de Tempo de Envio')
plt.xlabel('Tempo de Envio em Dias')
plt.ylabel('FrequÃªncia')
plt.show
```

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x = df['Time'])
plt.title('Boxplot Tempo de Envio')
plt.xlabel('Tempo de Envio em Dias')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

venda_mes = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum().reset_index()
venda_mes['Order Date'] = venda_mes['Order Date'].dt.to_timestamp()
venda_mes
```

```python
plt.figure(figsize = (10, 6))
sns.lineplot(x = 'Order Date', y = 'Sales', data = venda_mes, marker = 'o')

plt.title('Vendas Mensais')
plt.xlabel('Data')
plt.ylabel('Vendas Totais')
plt.xticks(rotation = 45)
plt.show()
```
