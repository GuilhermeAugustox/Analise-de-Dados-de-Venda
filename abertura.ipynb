```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Postal Code'].fillna(df['Postal Code'].mean(), inplace=True)
```

```python
df.head(2)
```

```python
receita = df['Sales'].sum()
receita
```

```python
receita_estado = df.groupby('State')['Sales'].sum()
receita_estado
```

```python
plt.figure(figsize = (50, 18))
sns.barplot(x = receita_estado.index, y = receita_estado.values)
plt.title('Receita por Estado')
plt.xlabel('Estado')
plt.ylabel('Receita')
plt.show()
```

```python
receita_regiao = df.groupby('Region')['Sales'].sum()
receita_regiao
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = receita_regiao.index, y = receita_regiao.values)
plt.title('Receita por Região')
plt.xlabel('Estado')
plt.ylabel('Receita')
plt.show()
```

```python
categoria_venda = df.groupby('Category')['Sales'].sum()
categoria_venda
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = categoria_venda.index, y = categoria_venda.values)
plt.title('Receita por Região')
plt.xlabel('Estado')
plt.ylabel('Receita')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
venda_mes = df.groupby(df['Order Date']).dt.to_period('M')['Sales'].sum().reset_index()
venda_mes['Order Date'] = venda_mes['Order Date'].dt.to_timestamp()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```
