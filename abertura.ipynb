```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Postal Code'].fillna(df['Postal Code'].mean(), inplace=True)
```

```python
df.head(2)
```

```python
receita = df['Sales'].sum()
receita
```

```python
receita_estado = df.groupby('State')['Sales'].sum()
receita_estado
```

```python
plt.figure(figsize = (50, 18))
sns.barplot(x = receita_estado.index, y = receita_estado.values)
plt.title('Receita por Estado')
plt.xlabel('Estado')
plt.ylabel('Receita')
plt.show()
```

```python
receita_regiao = df.groupby('Region')['Sales'].sum()
receita_regiao
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = receita_regiao.index, y = receita_regiao.values)
plt.title('Receita por Região')
plt.xlabel('Estado')
plt.ylabel('Receita')
plt.show()
```
